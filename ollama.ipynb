```python
# Cell 1: Import libraries and setup
import ollama
import random
import json

# Define keywords for each rating
RATING_KEYWORDS = {
    1: ["terrible", "awful", "horrible", "worst", "disappointing", "poor", "unacceptable", "bad"],
    2: ["below average", "not good", "unsatisfactory", "lacking", "mediocre", "subpar", "inadequate"],
    3: ["average", "okay", "decent", "fair", "acceptable", "moderate", "reasonable", "satisfactory"],
    4: ["good", "nice", "enjoyable", "pleasant", "solid", "quality", "impressive", "worthwhile"],
    5: ["excellent", "amazing", "outstanding", "fantastic", "perfect", "exceptional", "superb", "wonderful"]
}

print("Setup complete! Keywords loaded for ratings 1-5")

```

```python
# Cell 2: Function to generate review using Ollama
def generate_review(rating, product_name="product"):
    """Generate a review based on rating using Ollama"""
    # Select random keywords for the given rating
    keywords = random.sample(RATING_KEYWORDS[rating], min(5, len(RATING_KEYWORDS[rating])))
    keyword_str = ", ".join(keywords)
    
    # Create prompt for Ollama
    prompt = f"""Write a short customer review (2-3 sentences) for a {product_name} with a rating of {rating}/5. 
    Use these keywords naturally in the review: {keyword_str}. 
    Make it sound like a real customer review."""
    
    response = ollama.chat(
        model='tinyllama',
        messages=[
            {
                'role': 'user',
                'content': prompt,
            },
        ]
    )
    
    return response['message']['content']

print("Function defined successfully!")
```

```python
# Cell 3: Get user input and generate review
rating = int(input("Enter your rating (1-5): "))

if rating < 1 or rating > 5:
    print("Invalid rating! Please enter a number between 1 and 5.")
else:
    product_name = input("Enter product/service name (optional, press Enter to skip): ") or "product"
    
    print(f"\nGenerating review for rating {rating}/5...")
    print(f"Selected keywords: {random.sample(RATING_KEYWORDS[rating], min(3, len(RATING_KEYWORDS[rating])))}")
    
    review = generate_review(rating, product_name)
    
    print(f"Rating: {rating}/5")
    print(f"Review:\n{review}")



    # Cell 4: Generate multiple reviews and save to JSON (optional)
def generate_multiple_reviews(num_reviews=5):
    """Generate multiple reviews with random ratings"""
    reviews = []
    
    for i in range(num_reviews):
        rating = random.randint(1, 5)
        review_text = generate_review(rating)
        
        reviews.append({
            "id": i + 1,
            "rating": rating,
            "review": review_text,
            "keywords": random.sample(RATING_KEYWORDS[rating], min(3, len(RATING_KEYWORDS[rating])))
        })
        
        print(f"Generated review {i+1}/{num_reviews} (Rating: {rating}/5)")
    
    # Save to JSON file
    with open('reviews.json', 'w') as f:
        json.dump(reviews, f, indent=2)
    
    print(f"\nAll {num_reviews} reviews saved to reviews.json")
    return reviews
```
